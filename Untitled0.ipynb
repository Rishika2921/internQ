```python
import sys
print(f"Python version: {sys.version}")


import pandas as pd
import numpy as np
import sklearn
import subprocess
import platform

print("✅ pandas version:", pd.__version__)
print("✅ numpy version:", np.__version__)
print("✅ scikit-learn version:", sklearn.__version__)


print("\n📋 System Information")
print(f"OS: {platform.system()} {platform.release()}")
print(f"Machine: {platform.machine()}")
print(f"Processor: {platform.processor()}")


result = subprocess.run(["git", "--version"], capture_output=True, text=True)
if result.returncode == 0:
    print(f"\n✅ Git installed: {result.stdout.strip()}")
else:
    print("❌ Git not found. Please install Git.")


name = subprocess.run(["git", "config", "user.name"], capture_output=True, text=True)
email = subprocess.run(["git", "config", "user.email"], capture_output=True, text=True)
print(f"👤 Git user: {name.stdout.strip()}")
print(f"📧 Git email: {email.stdout.strip()}")


# import os
# os.makedirs("my_internship_project", exist_ok=True)
# os.chdir("my_internship_project")
# subprocess.run(["git", "init"])


readme_content = """# Internship Project - Week 01

## Developer Workspace & Version Control Setup

### Environment
- Python: {}
- Pandas: {}
- NumPy: {}
- Scikit-Learn: {}

### Setup Steps
1. Installed Python and Jupyter Notebook
2. Installed required libraries: pandas, numpy, scikit-learn
3. Verified Git installation and configuration
4. Created GitHub repository

### Status: ✅ Complete
""".format(sys.version.split()[0], pd.__version__, np.__version__, sklearn.__version__)

# Save README
with open("README.md", "w") as f:
    f.write(readme_content)

print("✅ README.md created successfully!")
print("\n📄 README Preview:")
print(readme_content)

#
print("=" * 50)
print("      SETUP VERIFICATION SUMMARY")
print("=" * 50)
checks = {
    "Python installed":        True,
    "Pandas installed":        True,
    "NumPy installed":         True,
    "Scikit-Learn installed":  True,
    "Git installed":           result.returncode == 0,
    "README.md created":       True,
}
for check, status in checks.items():
    icon = "✅" if status else "❌"
    print(f"{icon} {check}")

print("=" * 50)
print("🎉 Week 01 Setup Complete! Ready to submit.")
```
